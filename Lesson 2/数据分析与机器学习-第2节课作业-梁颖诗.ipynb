# 1. 从fundemantals.csv开始！

fundemantals.csv 是这些股票的年报数据

请用数据回答以下问题：

1. S&P500股票在2015年`net income`的均值是多少？最大值比最小值多多少？（每问10分，共计20分）
2. S&P500股票在2016年的固定资产（fixed assets）占总资产(total assets)比例的均值是多少？固定资产占总资产比例最小的股票是的代码（ticker symbol）是什么？（每问10分，共计20分）


```python
fd = pd.read_csv('fundamentals.csv')
fd.head()
```

```python
# 第1题 股票在2015年`net income`的均值是多少？最大值比最小值多多少？
fd2015 = fd[fd['For Year'] == 2015.0]
# fd2015.head()
print('S&P500股票在2015年net income的均值是',fd2015['Net Income'].mean())
print('S&P500股票在2015年最大值是',fd2015['Net Income'].max())
print('S&P500股票在2015年最小值是',fd2015['Net Income'].min())
print('S&P500股票在2015年最大值比最小值多',fd2015['Net Income'].max() - fd2015['Net Income'].min()) 
```

```python
# 第2题 股票在2016年的固定资产（fixed assets）占总资产(total assets)比例的均值是多少？
#   固定资产占总资产比例最小的股票的代码（ticker symbol）是什么？
fd2016 = fd[fd['For Year'] == 2016.0]
# fd2016.head()
fd2016.loc[:,'fixed assets/total assets'] = fd2016.loc[:,'Fixed Assets']/fd2016.loc[:,'Total Assets']
# fd2016.head()
print('股票在2016年的固定资产（fixed assets）占总资产(total assets)比例的均值是',fd2016['fixed assets/total assets'].mean())
a = fd2016['fixed assets/total assets'].min()
fd2016_1 = fd2016[['Ticker Symbol','fixed assets/total assets']]
fd2016_1.head()
fd2016_2 = fd2016_1[fd2016_1['fixed assets/total assets'] == a]
fd2016_2.head()
print('固定资产占总资产比例最小的股票的代码（ticker symbol）是',fd2016_2['Ticker Symbol'])
```

# 2. 加入securities.csv~

securities.csv包含了这些股票的基本信息

1. 请列举出各个sector中的加入时间最早的股票名称（10分）
2. 请列举出每一个州中加入时间最晚的股票名称（10分）


```python
# 第1题 请列举出各个sector中的加入时间最早的股票名称
se = pd.read_csv('securities.csv')
se = se.dropna(subset=["Date first added"])
se['Date first added'] = pd.to_datetime(se['Date first added'])
#se.head()
se1 = se.sort_values(by='Date first added')
#se1.head()
se2 = se1.groupby('GICS Sector')['Date first added'].min()
print('各个sector中的加入时间最早的股票名称\n',se2)
```

```python
# 第2题 每一个州中加入时间最晚的股票名称
#se.head()
se3 = se1.groupby('Address of Headquarters')['Date first added'].max()
print('每一个州中加入时间最晚的股票名称\n',se3)
```

现在你需要同时处理来自两个表中的信息了

1. 请思考，合并两个表的信息的时候，我们应该用什么样的准则对其它们（10分）
2. 请列举每个sector在2013-2016年累计Research&Development的总投入（10分）
3. 请列举出每个sector中，在2013-2016年累计Research&development投入最大的3家公司的名称以及投入的数值（20分）

```python
# 第1题
se = se.rename(columns={'Ticker symbol':'Ticker Symbol'}) # 列名需相同，区分大小写
se.head()
```

```python
#se.head()
mer = pd.merge(fd, se, how = 'inner',on = ['Ticker Symbol'])
mer.head()
```

```python
#第2题：列举每个sector在2013-2016年累计Research&Development的总投入
m2 = mer[(mer['For Year'] >=2013.0) & (mer['For Year'] <=2016.0)]
#m2.head(10)
m2_2 = m2.groupby('GICS Sector')['Research and Development'].sum()
print('每个sector在2013-2016年累计Research&Development的总投入',m2_2)
```

```python
#第3题： 请列举出每个sector中，在2013-2016年累计Research&development投入最大的3家公司的名称以及投入的数值
m3 = m2.sort_values(by = 'Research and Development')
m3 = m3.groupby(['GICS Sector']).head(3)
#m3.head()
print(m3)
```
